### Copy the users and roles DataFrames from the examples above.

```python
users = pd.DataFrame({
    'id': [1, 2, 3, 4, 5, 6],
    'name': ['bob', 'joe', 'sally', 'adam', 'jane', 'mike'],
    'role_id': [1, 2, 3, 3, np.nan, np.nan]
})
users
```

```python
roles = pd.DataFrame({
    'id': [1, 2, 3, 4],
    'name': ['admin', 'author', 'reviewer', 'commenter']
})
roles
```

### What is the result of using a right join on the DataFrames?

```python
pd.merge(users,roles, on = "id", how = "right")
```

### What is the result of using an outer join on the DataFrames?

```python
pd.merge(users,roles,on = "id", how = "outer")
```

### What happens if you drop the foreign keys from the DataFrames and try to merge them?

```python
users_no_foreign_key = users.loc[:,"id":"name"]
```

```python
users_no_foreign_key
```

```python
roles
```

```python
outer_join = pd.merge(users_no_foreign_key,roles, on = "id", how = "outer")
outer_join.rename(columns = {"name_x": "name","name_y":"role"})
```

```python
inner_join = pd.merge(users_no_foreign_key,roles, on = "id", how = "inner")
inner_join.rename(columns = {"name_x": "name","name_y":"role"})
```

### Create a column named mileage_difference like you did in the DataFrames exercises; this column should contain the difference between highway and city mileage for each car.

```python
mpg = mpg.assign(mileage_difference = (mpg.highway - mpg.city))
mpg
```

### Create a column named average_mileage like you did in the DataFrames exercises; this is the mean of the city and highway mileage.

```python
mpg = mpg.assign(average_mileage = (mpg.city + mpg.highway) / 2)
mpg
```

### Create a new column on the mpg dataset named is_automatic that holds boolean values denoting whether the car has an automatic transmission.

```python
mpg["is_automatic"] = np.where((mpg["trans"] == "manual(m5)") | (mpg["trans"] == "manual(m6)"), False, True)
```

```python
mpg["is_automatic"]
```

```python
#This way reads nicer below
```

```python
mpg["is_automatic"] = np.where((mpg["trans"] == "manual(m5)") | (mpg["trans"] == "manual(m6)"), "manual", "automatic") 
mpg
```

### Use your get_db_url function to help you explore the data from the chipotle database.

```python
chipotle_url = get_db_url(user,password,host,database = "chipotle")
```

```python
query = '''
SELECT *
FROM orders;
'''
chipotle = pd.read_sql(query,chipotle_url)
chipotle
```

### What is the total price for each order?

```python
chipotle["item_price"] = chipotle["item_price"].str.lstrip("$")
```

```python
chipotle
```

```python
chipotle = chipotle.astype({"item_price":"float64"})
```

```python
chipotle["order_total"] = chipotle['item_price'] * chipotle['quantity']
```

```python
chipotle.groupby("order_id").order_total.sum()
```

### What are the most popular 3 items?

```python
chipotle
```

```python
chipotle.groupby("item_name").sum().sort_values(by=["quantity"], ascending = False).head(3)
```

```python
# chicken bowl, chicken burrito, chips and guac
```

### Which item has produced the most revenue?

```python
chipotle.groupby("item_name").sum().sort_values(by="item_price", ascending = False).head(1)
```

### Write the code necessary to create a cross tabulation of the number of titles by department. (Hint: this will involve a combination of SQL code to pull the necessary data and python/pandas code to perform the manipulations.)

```python
titles
```

```python
employees
```

```python
query = '''
SELECT emp_no,dept_no
FROM dept_emp;
'''
dept_emp = pd.read_sql(query,url)
dept_emp
```

```python
finding_titles = emp_with_titles.merge(dept_emp, on = "emp_no", how = "outer")
```

```python
df = finding_titles.groupby(["dept_no","title"]).count()
```

```python
df[df.columns[0]].count("dept_no")
```

```python
pd.crosstab(finding_titles.title,finding_titles.dept_no)
```
