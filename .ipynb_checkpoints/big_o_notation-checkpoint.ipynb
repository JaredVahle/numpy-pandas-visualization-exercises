```python
import matplotlib.pyplot as plt
import math
```

```python
n = list(range(1,7))
```

```python
plt.plot(n)
```

```python
plt.plot(n,[0*y+1 for y in n])
plt.plot(n,[math.log(y) for y in n])
plt.plot(n,[y for y in n])
plt.plot(n,[y * math.log(y) for y in n])
plt.plot(n,[y**2 for y in n])
plt.plot(n,[2**y for y in n])
```
