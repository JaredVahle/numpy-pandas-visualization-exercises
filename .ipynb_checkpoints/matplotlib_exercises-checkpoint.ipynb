### Use matplotlib to plot the following equation:

y = x ^ 2 - x + 2

### You'll need to write the code that generates the x and y points.

```python
i = [x for x in range(-10,50)]
```

```python
len(i)
```

```python
def x_squared_minus_x_plus_two(i):
    y = []
    for x in i:
        y.append((x**2) - x + 2)
    return y
```

```python
y = x_squared_minus_x_plus_two(i)
```

```python
len(y)
```

### Add an anotation for the point 0, 0, the origin.

```python
plt.plot(i,y)
plt.text(0.25, 50, '(0, 0)', fontsize=10, color='blue')
```
